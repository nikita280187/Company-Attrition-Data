#Removing outliers for precise data

```python
sns.set(style='whitegrid')
columns_to_boxplot = ['white_rating', 'black_rating']
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[columns_to_boxplot])
plt.title('Box Plot of Ratings')
plt.ylabel('Rating')
plt.show()
```

```python
def remove_outliers(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[column] >= lower_bound) & (df[column]<= upper_bound)]
    return df
```

```python
clean_data = remove_outliers(df, columns_to_boxplot)
plt.figure(figsize=(10, 6))
sns.boxplot(data=clean_data[columns_to_boxplot])
plt.title('Box Plot of Ratings (Outliers Removed)')
plt.ylabel('Rating')
plt.show()
```
